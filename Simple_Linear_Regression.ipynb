##**Simple Linear Regression**

**Formula for Simple Linear Regression**  
**ŷ = b₀ + b₁X₁**  
**ŷ:** Dependent Variable  
**b₀:** y-intercept(constant)  
**b₁X₁:** Independent Variable  

**Example:** Amount of Potatoes produced based on the type of Fertilizer we use.  
**Potatoes[*t*] = b₀ + b₁ x Fertilizer[*kg*]**

###**Importing the Libraries**

```python
# Importing the required Libraries like pandas, numpy and matplotlib.pyplot
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

###**Importing the Dataset**

```python
# Import the required Dataset and View It
dataset = pd.read_csv("/content/Salary_Data.csv")
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

###**Splitting Dataset into Training and Test Set**

```python
# Splitting the Dataset into Training and Test Set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

###**Training the Simple Linear Regression Model on the Training Set**

```python
# Building a Simple Linear Regression Model
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
# fit() method will train the Regression Model in the Training Set
regressor.fit(X_train, y_train)
```

###**Predicting the Test Set Results**

```python
# Predicting the Results from Test Set
y_pred = regressor.predict(X_test)
```

###**Visualizing the Training Set Results**

```python
# Plotting a 2D Graph for Visualization
plt.scatter(X_train, y_train, color = 'red')
plt.plot(X_train, regressor.predict(X_train), color = 'blue')
plt.title('Salary vs Experience (Training Set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

###**Visualizing the Test Set Results**

```python
# Plotting a 2D Graph for Visualization
plt.scatter(X_test, y_test, color = 'red')
'''
The Regression Line that we get is actually resulting from a Unique Equation and therefore the Predicted Salaries
of the Test Set will be on the same Regression Line as the Predicted Salaries of the Training Set
'''
# So we do not need to replace the X_train to X_test in the below Equation.
plt.plot(X_train, regressor.predict(X_train), color = 'blue')
plt.title('Salary vs Experience (Test Set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```
